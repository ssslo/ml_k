# Regression Task


### 1. Data  Preparation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('kc_house_data.csv')
data.head()
```

```python
print("Shape: ", data.shape)
data.describe()
```

```python
data.info()
```

```python
data.hist(figsize=(20,15),bins=50)
plt.show()
```

```python
def univariate_plot(data, figsize, index):
    n_rows, n_cols = index
    plt.figure(figsize=figsize)
    for idx, var in enumerate(data.columns, 1):
        plt.subplot(n_rows, n_cols, idx)
        sns.boxplot(x = data[var])
```

```python
univariate_plot(data, figsize=(14, 20), index=(4,4))
```

```python
def remove_outliers(data, var_names):
    ans = data.copy()
    for var in var_names:
        var_info = ans[var].describe()        
        iq_range = var_info['75%'] - var_info['25%']
        intv_range = (var_info['25%'] - 1.5 * iq_range, var_info['75%'] + 1.5 * iq_range)
        ans = ans[ans[var].between(*intv_range)]
    return ans
```

```python
new_data = remove_outliers(data, data.columns)
univariate_plot(new_data, figsize=(14, 20), index=(4,4))
```

```python
new_data['price'] = np.log(new_data['price'])
```

```python
new_data['yr_renovated'] = new_data[['yr_renovated', 'yr_built']].apply(lambda pair: pair[0] if pair[0] != 0 else pair[1], axis=1)
```

```python
train_set, test_set = train_test_split(new_data, test_size=0.2, random_state=42)
X_train = train_set.drop('price', axis=1)
y_train = train_set['price'].copy()
X_test = test_set.drop('price', axis=1)
y_test = test_set['price'].copy()
```

### 2.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
def pred(model, X):
    return model.predict(X)
def rms(y, y_pred):
    return np.sqrt(mean_squared_error(y, y_pred))
def accuracy(model, X, y):
    return model.score(X, y)
```

```python
scaler = StandardScaler()
X_train_prep = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test_prep = pd.DataFrame(scaler.fit_transform(X_test), columns=X_test.columns)
```

```python
lin_reg = LinearRegression()
lin_model = lin_reg.fit(X_train_prep, y_train)
```

```python
print("Prediction Train: ", accuracy(lin_reg, X_train_prep, y_train))
print("Prediction Test: ", accuracy(lin_reg, X_test_prep, y_test))
```

# Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree_reg = DecisionTreeRegressor()
tree_model = tree_reg.fit(X_train_prep, y_train)
```

```python
print("Prediction Train: ", accuracy(tree_reg, X_train_prep, y_train))
print("Prediction Test: ", accuracy(tree_reg, X_test_prep, y_test))
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_reg = RandomForestRegressor()
rf_model = rf_reg.fit(X_train_prep, y_train)
```

```python
print("Prediction Train: ", accuracy(rf_reg, X_train_prep, y_train))
print("Prediction Test: ", accuracy(rf_reg, X_test_prep, y_test))
```

#  Neural Networks

```python
from tensorflow import keras
from keras.layers import Dense
from keras.models import Sequential
```

```python
n_layers = 7
n_units = X_train.shape[1]

model = Sequential()
for i in range(n_layers) :
    model.add(Dense(n_units, activation="relu"))
    
keras.layers.Dense(1, activation="linear")
model.compile(loss="mse", optimizer="adam")
```

```python

history = model.fit(X_train_prep, y_train, epochs=500 , batch_size=100, validation_data=(X_test, y_test), verbose=False)
```

```python
pd.DataFrame(history.history).plot(figsize=(8, 5))
plt.grid(True)
plt.show()
```

```python
mse_test = model.evaluate(X_test_prep, y_test)
```

```python
y_train_pred = model.predict(x=X_train_prep)
y_test_pred = model.predict(x=X_test_prep)
```
